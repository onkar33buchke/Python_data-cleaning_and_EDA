```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('glassdoor_jobs.csv')

df.head(3)
```

#### Remove rows where salary estimate value is equal to '-1'

```python
df = df[df['Salary Estimate'] != '-1']

df.head(3)
```

#### Salary parsing

```python
# Remove text between parenthesis (eg.("Glassdoor est.")) from 'Salary Estimate' column using split
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary.head(2)
```

```python
# Remove "$" and "K" from 'Salary Estimate' column

minus_KD = salary.apply(lambda x: x.replace('K',"").replace("$",''))
minus_KD.head(2)
```

```python
# Identifying rows where "Salary Estimate" is in Hour 

df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
# Remove text "per hour" and "employer proved salary:" from 'Salary Estimate' column

min_hr = minus_KD.apply(lambda x: x.lower().replace('per hour',"").replace("employer provided salary:",''))
```

```python
# Spliting salary range as 'min_salary' and 'max_salary' using "-"

df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
```

```python
df['avg_salary'] = (df.min_salary + df.max_salary)/2
```

#### Extracting only company name from 'Company Name' column

```python
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3],axis=1)

df.head(3)
```

#### Extract state field from 'Location' column

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])

df.head(2)
```

```python
df.job_state.value_counts().head(5)
```

#### Finding the job positions that are at the headquarters

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0,axis=1)

df.head(3)
```

#### Finding age of the company

```python
df['age'] = df.Founded.apply(lambda x: x if x < 1 else 2021 - x)
df.head(3)
```

#### Parsing of job description (python,etc.)

```python
# python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)

# r studtio

df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower () else 0)

# spark

df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() in x.lower () else 0)

# aws

df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() in x.lower () else 0)

# excel

df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() in x.lower () else 0)
```

```python
df_out = df.drop(['Unnamed: 0'],axis=1)

df_out.head(3)
```

```python
df_out.to_csv("salary_data_cleaned.csv",index=False)
```
